```python
unique_uuid = results.uuid.unique()
user_login_asset_final = pd.DataFrame()
for u in unique_uuid:
    user = results[results.uuid==u]
#     print(user.head())
    user_token_auth = user.user_token_auth.unique()
    for l in user_token_auth:
        user_login = user[user.user_token_auth==l]
        user_login_asset = user_login.assetId.unique()
        for a in user_login_asset:
            user_login_asset = user_login[user_login.assetId==a]
            user_login_asset.sort_values(by = 'time')
            user_login_asset['playback_time'] = 0
            user_login_asset.reset_index()
            user_login_asset['index1'] = user_login_asset.index
            print(user_login_asset.head())
            for i in range(len(user_login_asset)-1):
                for j in range(len(combi_df_yn)):
                    if (user_login_asset.iloc[i]['action'] == combi_df_yn.iloc[j]['event1']) and (user_login_asset.iloc[i+1]['action']==combi_df_yn.iloc[j]['event2']) and (combi_df_yn.iloc[j]['played_yn']=='y'):
                        if user_login_asset.iloc[i+1]['action']!= 'stop' or 'Stop' or 'close' or 'finish':
                            user_login_asset.at[(user_login_asset.iloc[i]['index1']),'playback_time'] = (user_login_asset.iloc[i+1]['sec_difference']-user_login_asset.iloc[i]['sec_difference'])            
                    if user_login_asset.iloc[i+1]['action'] == 'stop' or 'Stop' or 'close' or 'finish':
                        user_login_asset.at[(user_login_asset.iloc[i]['index1']),'playback_time'] = (user_login_asset.iloc[i+1]['timestamp']-user_login_asset.iloc[i]['timestamp'])
                        if user_login_asset_final.empty == True:
                            user_login_asset_final = user_login_asset
                        else:
                            user_login_asset_final.append(user_login_asset)
# user_login_asset_final = pd.DataFrame(columns=user_login_asset.columns)       
print(user_login_asset_final.head())
```
