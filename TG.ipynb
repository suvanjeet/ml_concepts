```python
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n']
```

```python
tokens = separate_punc(text)
```

```python
tokens_c = [x.replace('\n', '') for x in tokens]
```

```python
# organize into sequences of tokens
train_len = 10+1 # 10 training words , then one target word

# Empty list of sequences
text_sequences = []

for i in range(train_len, len(tokens_c)):
    
    # Grab train_len# amount of characters
    seq = tokens_c[i-train_len:i]
    
    # Add to list of sequences
    text_sequences.append(seq)
```

```python
# integer encode sequences of words
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
def create_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, 100, input_length=seq_len))
    model.add(LSTM(150, return_sequences=True))
    model.add(LSTM(150))
    model.add(Dense(150, activation='relu'))

    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model
```

```python
model = create_model(vocabulary_size+1, 10)
```

```python
model.fit(X, y, batch_size=64, epochs=30,verbose=1)
```
